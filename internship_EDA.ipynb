**STEP6**: **OUTLIERS** **TREATMENT**

for cont numerical data only not categorical and **discrete** mor outliers no need to do

polio is discrete so no need

````python
def wisker(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3-q1
  lw=q1-1.5*iqr
  uw=q3+1.5*iqr
  return lw,uw
````

````python
wisker(df['GDP'])
````

lower whisker and upper whisker

````python
for i in ['GDP','Total expenditure',' thinness  1-19 years',' thinness 5-9 years']: # Added spaces to match the actual column names
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])

````

**IMPORTING** **LIBRARIES**

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
````

**READING** **THE** **DATA**

````python
df = pd.read_csv('/content/archive (2) (1).zip')
````

````python


# Drop rows with missing values (or use fillna if preferred)
df = df.dropna()

# Drop irrelevant columns and one-hot encode categorical features
X = df.drop(columns=['Life expectancy ', 'Country', 'Year'])
X = pd.get_dummies(X, columns=['Status'], drop_first=True) # One-hot encode 'Status'
y = df['Life expectancy ']

# Ensure X and y have the same number of samples before splitting
X = X.loc[y.index]

# Now proceed with train_test_split


````

**SPLITING** **THE** **DATA** **INTO** **TRAIN** AND **TEST**

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=101
)
````

**TRAINING** **THE** **MODEL**

````python
# Create and train the model
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
````

TESTING **THE** **MODEL**

````python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
y_pred = np.array(y_pred)
y_test = np.array(y_test)

print(np.concatenate((y_pred.reshape(len(y_test),1), y_test.reshape(len(y_test),1)),1))
````

**ACCURACY** **OF** **THE** **MODEL**

````python
accuracy_score =  regressor.score(X_test,y_test)
print(accuracy_score)
````

````python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(regressor,X_train,y_train,cv=10)
accuracies.mean()
````
